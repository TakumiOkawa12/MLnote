# クラスタリング

## K-means

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')

X = df.drop(columns = ["species"])

# 標準化
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# k-means
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, random_state = 0)
kmeans.fit(X_scaled)
```

```python
label = kmeans.predict(X_scaled)
label
```

```python
import numpy as np
np.std(label)
```

＜チャレンジ＞

```python
import pandas as pd
label_df = pd.DataFrame(label, columns = ["label"])
label_df["true_label"] = label_df["label"].map({0:"setosa", 1:"versicolor", 2:"virginica"})
temp_df = pd.concat([df.iloc[:, :4], label_df["true_label"]], axis = 1)
sns.pairplot(temp_df, hue = "true_label")
```

```python
sns.pairplot(df, hue = "species")
```

```python
X_scaled = StandardScaler().fit_transform(X)
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(X_scaled)
```

＜動画内回答＞

```python
# データ準備
result_df = pd.concat([X, pd.DataFrame(label, columns = ["label"])], axis = 1)
sns.pairplot(result_df, hue = "label")
```

```python
sns.pairplot(df, hue = "species")
```

```python
losses = []
for K in range(1, 10):
    kmeans = KMeans(n_clusters = K, random_state = 0).fit(X_scaled)
    # scoreは値が高い方が良いとなっているので、マイナスをつける(誤差関数は値が小さい方が良い)
    losses.append(-kmeans.score(X_scaled))
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 10), losses)
plt.xlabel('K')
plt.ylabel('loss')
```

## 階層クラスタリング

```python
from scipy.cluster.hierarchy import linkage, dendrogram
Z = linkage(X_scaled, method = 'ward')
```

```python
print(X_scaled.shape)
print(Z.shape)
```

```python
Z[51]
```

```python
# 159番目のデータは存在しない
# 159は159-150(X_scaledのデータ数)= 9番目を表す
# つまりZの9番目を見ると、下の階層のデータのインデックスが分かる
```

```python
Z[159-len(X_scaled)]
```

```python
X_scaled[19]
```

```python
d = dendrogram(Z, truncate_mode = 'lastp', p = 3)
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
# Zをラベル付けする
label_cluster = fcluster(Z, criterion = 'maxclust', t = 3)
```

```python
cluster_df = pd.concat([pd.DataFrame(X_scaled, columns = X.columns), pd.DataFrame(label_cluster, columns = ["label_cluster"])], axis = 1)
sns.pairplot(cluster_df, hue = "label_cluster")
```
