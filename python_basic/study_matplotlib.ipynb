```python
import matplotlib.pyplot as plt
figure_ = plt.figure(1)  # Figure作成
axes_ = figure_.add_subplot(111)  # Axes作成
xaxis_ = axes_.xaxis  # XAxis取得
xaxis_.get_label()  # X axis label取得
```

- 「figure 1」 という図のオブジェクトを作成している
- そのオブジェクトに.add_subplot()で何行何列の何番目のaxesのオブジェクトを作成している
- そしてそのオブジェクトに対してplotしている

```python
# sample data
import numpy as np
x = np.arange(0, 10 ,1)
y = np.power(x, 2)

# Figure and Axes
import matplotlib.pyplot as plt
plt.close(1)  # 既にFigure1が開かれていれば閉じる
figure_ = plt.figure(1, figsize=(8,4))
axes_ = figure_.add_subplot(121)
axes_.plot(x, y, 'o-')
axes__ = figure_.add_subplot(122)
axes__.plot(x, y, 'o-')
```
